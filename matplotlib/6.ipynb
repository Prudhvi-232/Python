```python
%matplotlib inline 
import matplotlib.pyplot as plt
```

```python
vals=[122,34,44,123]
names=["homw","lawda","hwllo","bye"]
plt.pie(vals,labels=names,shadow=True,autopct='%1.1f%%',radius=2)
plt.axis("equal")
plt.show()
```

```python
plt.axis("equal")
plt.pie(vals,labels=names, shadow=True, autopct='%1.1f%%',radius=1.5,explode=[0,0.03,0.06,0.09])
plt.show()
```
