```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
days = [1, 2, 3, 4, 5, 6]      # Example: Length = 6
max_t = [30, 32, 31, 29, 28, 27]  # Example: Length = 6
min_t = [15, 17, 16, 14, 13, 12]  # Example: Length = 6
avg_t = [22, 24, 23, 21, 20, 19]  # Example: Length = 6
```

```python
plt.plot(days, max_t, label="max")
plt.plot(days, min_t, label="min")
plt.plot(days, avg_t, label="average")
```
